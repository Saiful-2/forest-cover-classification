## COMPLETE EDA + INSIGHTS PROJECT
#### STEP 0: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.set_style("whitegrid")
```

## STEP 1: Load Clean Dataset

```python
url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/forest_cleaned_dataset.csv"
df = pd.read_csv(url)

df.head()
```

## STEP 2: Basic Overview

```python
df.shape
df.info()
df.describe()
```

## STEP 3: Target Variable Analysis

```python
sns.countplot(x='Cover_Type', data=df)
plt.title("Distribution of Forest Cover Types")
plt.show()
```

### 🧠 Insight:
Some cover types dominate → dataset may be imbalanced
Important for machine learning

## STEP 4: Univariate Analysis

```python
sns.histplot(df['Elevation'], kde=True)
plt.title("Elevation Distribution")
plt.show()
```

#### 🧠 Insight:
Check if normally distributed or skewed

### Slope Distribution

```python
sns.histplot(df['Slope'], kde=True)
plt.title("Slope Distribution")
plt.show()
```

### 📌 Aspect Distribution

```python
sns.histplot(df['Aspect'], kde=True)
plt.title("Aspect Distribution")
plt.show()
```

### Elevation vs Cover Type

```python
sns.boxplot(x='Cover_Type', y='Elevation', data=df)
plt.title("Elevation vs Cover Type")
plt.show()
```

### Slope vs Cover Type

```python
sns.boxplot(x='Cover_Type', y='Slope', data=df)
plt.title("Slope vs Cover Type")
plt.show()
```

### Aspect vs Cover Type

```python
sns.boxplot(x='Cover_Type', y='Aspect', data=df)
plt.title("Aspect vs Cover Type")
plt.show()
```

## STEP 6: Correlation Analysis

```python
corr_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## STEP 7: Feature Relationships

```python
sns.pairplot(df[['Elevation','Slope','Aspect','Cover_Type']], hue='Cover_Type')
```

## Insight
1. Elevation is the strongest feature, showing clear separation between different cover types.
2. Slope and Aspect are weak individually, as their values overlap heavily across classes.
3. The dataset shows significant class overlap, meaning simple models won’t perform well.
4. Most features are not strongly correlated, which is good for machine learning models.
5. Overall, tree-based models (like Random Forest) are suitable for capturing these complex patterns.

## STEP 8: Advanced Insight

```python
df.groupby('Cover_Type')[['Elevation','Slope']].mean()
```

## STEP 9: Final Key Insights

Elevation plays a major role in determining forest cover type

Slope has relatively low influence on cover type
