```python
import pandas as pd

url = "https://raw.githubusercontent.com/Saiful-2/datasets/main/forest_dirty_dataset.csv"
df = pd.read_csv(url)

df.head()
```

```python
soil_cols = [col for col in df.columns if "Soil_Type" in col]

df['Soil_Sum'] = df[soil_cols].sum(axis=1)
df['Soil_Sum'].value_counts()
```

## STEP 4: Fix Missing Values

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)
```

## STEP 5: Remove Duplicates

```python
df = df.drop_duplicates()
```

## STEP 6: Handle Outliers

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
cols = ['Elevation', 'Slope', 'Horizontal_Distance_To_Roadways']

for col in cols:
    df = remove_outliers(df, col)
```

## STEP 7: Fix Validation Errors

```python
def fix_one_hot(row):
    if row[soil_cols].sum() != 1:
        max_col = row[soil_cols].idxmax()
        row[soil_cols] = 0
        row[max_col] = 1
    return row

df = df.apply(fix_one_hot, axis=1)
```

## STEP 8: Verify Everything
